```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_excel("/content/Book1.xlsx")

read_y=df["TEMPERATURE"]
read_x=df["DAY"]
mean_value=read_y.mean()
min_value=read_y.min()
max_value=read_y.max()
print("MEAN TEMPERATURE=",mean_value)
print("MINIMUM TEMPERATURE=",min_value)
print("MAXIMUM TEMPERATURE=",max_value)
plt.subplot(2,1,1)
plt.title("TEMPERATURE IN A MONTH")
plt.plot(read_x,read_y,color="BLUE",linewidth='2',label="tempartures",marker='.')
plt.legend(loc="best")
plt.subplot(2,1,2)


threshold_value=read_y[read_y>mean_value]
print("The threshold days where the temperatures exceeded mean are",len(threshold_value))
plt.plot(threshold_value.index,threshold_value,color="red",linewidth='2',label="temperatures above mean",marker=".")
plt.subplot(2,1,2)
plt.legend(loc="lower right")
plt.show()
```
