```python
import seaborn as sns
import matplotlib.pyplot as plt


tips=sns.load_dataset("tips")
sns.scatterplot(x="total_bill",y="tip",data=tips)
plt.title("Scatter plot od Total Bill vs.Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


tips=sns.load_dataset("iris")
sns.scatterplot(x="sepal_length",y="sepal_width",data=tips)
plt.title("Scatter plot of Sepal length and sepal width")
plt.xlabel("sepal_length")
plt.ylabel("sepal_width")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


titanic = sns.load_dataset("titanic")

# Create a scatterplot
sns.scatterplot(x="age", y="fare", data=titanic)

plt.title("Scatterplot of Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")

# Display the plot
plt.show()
```

```python
import seaborn as sns

iris=sns.load_dataset("iris")

correlation_matrix=iris.corr()

sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap of Brain_dataset Dataset")
plt.show()
```
